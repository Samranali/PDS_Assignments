```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
raw_data = pd.read_csv('raw_data/train.csv').iloc[:, 1:]
```

```python
missing_values = raw_data.isnull().sum()
```

```python
raw_data = raw_data.apply(lambda col: col.fillna(col.mode()[0]) if col.dtype == 'object' else col.fillna(col.median()))
```

```python
if 'New_Price' in raw_data.columns:
    raw_data.drop(columns=['New_Price'], inplace=True)
```

```python
missing_values_after = raw_data.isnull().sum()
```

```python
print("Missing values before processing:\n", missing_values)
```

```python
print("\nMissing values after processing:\n", missing_values_after)
```

```python
df=raw_data
df.head()
```

```python
if 'Mileage' in raw_data.columns:
    raw_data['Mileage'] = raw_data['Mileage'].str.extract('(\d+.\d+)').astype(float)

df.head()
```

```python
if 'Engine' in raw_data.columns:
    raw_data['Engine'] = raw_data['Engine'].str.extract('(\d+)').astype(float)

df.head()
```

```python
if 'Power' in raw_data.columns:
    raw_data['Power'] = raw_data['Power'].str.extract('(\d+.\d+)').astype(float)

df.head()
```

```python
if 'Fuel_Type' in raw_data.columns and 'Transmission' in raw_data.columns:
    raw_data = pd.get_dummies(raw_data, columns=['Fuel_Type', 'Transmission'], drop_first=True)

df.head()
```

```python
current_year = datetime.now().year
if 'Year' in raw_data.columns:
    raw_data['Car_Age'] = current_year - raw_data['Year']

print("\nColumns after adding 'Car_Age':\n", raw_data.columns)
```

```python
selected_data = raw_data[['Name', 'Year', 'Mileage', 'Engine', 'Power', 'Car_Age']]
print("Selected Data:\n", selected_data.head())
```

```python
filtered_data = selected_data[(selected_data['Mileage'] > 15) & (selected_data['Engine'] > 1000)]
print("\nFiltered Data:\n", filtered_data.head())
```

```python
renamed_data = filtered_data.rename(columns={'Car_Name': 'Model', 'Car_Age': 'Age'})
renamed_data['Mileage_per_Engine'] = renamed_data['Mileage'] / renamed_data['Engine']
print("\nRenamed Data:\n", renamed_data.head())
```

```python
arranged_data = renamed_data.sort_values(by='Mileage', ascending=False)
print("\nArranged Data:\n", arranged_data.head())
```

```python
summarized_data = raw_data.groupby('Year').agg(avg_mileage=('Mileage', 'mean'), avg_engine=('Engine', 'mean')).reset_index()
print("\nSummarized Data:\n", summarized_data.head())
```

```python
raw_data.to_csv('processed_data/processed.csv', index=False)
```
